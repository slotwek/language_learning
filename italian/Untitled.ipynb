```python
import requests
from bs4 import BeautifulSoup

# Specify the URL of the website you want to scrape
url = 'http://example.com'
url = 'https://easyitaliannews.com/'

# To avoid server error: 403
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/117.0"
}

# Send a GET request to the URL
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Now you can extract data from the soup object
    # For example, to get the title of the page
    title = soup.title.string
    print(f"Page Title: {title}")
    
    # Extracting all paragraphs
    paragraphs = soup.find_all('p')
    for idx, paragraph in enumerate(paragraphs):
        print(f"Paragraph {idx+1}: {paragraph.text}")
    
    # Extracting all links
    links = soup.find_all('a')
    for link in links:
        print(f"Link: {link.get('href')} Text: {link.text}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
ns = soup.find('div', {'class': 'entry-content'})
```

```python
# Extracting all paragraphs
l_content = []
paragraphs = ns.find_all(True)
for idx, paragraph in enumerate(paragraphs):
    print(f"Paragraph {idx+1}: {paragraph.text}")
    p = str(paragraph.text)
    if  p not in l_content:
        l_content.append(p)
        
```

```python
l_content
```

```python
for el in l_content:
    try:
        if 'www' in el:
            print(el)
            l_content.remove(el)
        elif '.com' in el:
            l_content.remove(el)
        elif '.it' in el:
            l_content.remove(el)

        if 'GovernmentZA' in el:
            l_content.remove(el)

        if 'Attribution' in el:
            l_content.remove(el)

        if 'Generic' in el:
            l_content.remove(el)
        if 'Click the link' in el:
            l_content.remove(el)
    except:
        pass
```

```python
s = ('\n').join(l_content).split('Il tuo aiuto per noi è importante!')[0]
```
